Задание по Github (вывод списка репозиториев)

```python
# Необходимые импорты
import requests
import json  

# Указываем token
token = token2

headers = {
    'Authorization': f'token {token}',
    'Accept': 'application/vnd.github+json'
}

url = 'https://api.github.com/user/repos?per_page=100'

all_repos = []
page = 1

while True:
    response = requests.get(f"{url}&page={page}", headers=headers)
    if response.status_code != 200:
        print(f"Ошибка запроса: {response.status_code} - {response.text}")
        break
    repos = response.json()
    if not repos:
        break
    all_repos.extend(repos)
    page += 1

# Сохраняем весь полученный JSON в файл
filename = 'my_repos_full.json'
with open(filename, 'w', encoding='utf-8') as f:
    json.dump(all_repos, f, ensure_ascii=False, indent=4)

print(f"Сохранено {len(all_repos)} репозиториев в файл '{filename}'.")

# Откроем сохраненный json с доступными репозиториями
with open(filename, 'r', encoding='utf-8') as f:
    content = f.read()

print(content)
```

Задание по VK (вывод списка групп)

```python
# Необходимые импорты
import urllib.request
import json
import pandas as pd  

# Указываем access_token и user_id
access_token = access_token2
user_id = user_id2  

# Формируем URL для запроса к VK API (метод groups.get)
params = {
    'v': '5.199',
    'access_token': access_token,
    'extended': 1,  # Возвращает дополнительные поля, такие как name и screen_name
    'fields': 'name,screen_name'  # Поля, которые хотим получить для каждой группы
}

if user_id:
    params['user_id'] = user_id

url = 'https://api.vk.com/method/groups.get?' + '&'.join(f"{k}={v}" for k, v in params.items())

# Выполняем запрос
try:
    with urllib.request.urlopen(url) as response:
        if response.getcode() == 200:
            data = json.loads(response.read().decode('utf-8'))
        else:
            print(f"Ошибка HTTP: {response.getcode()}")
            data = None
except Exception as e:
    print(f"Ошибка запроса: {e}")
    data = None

# Парсим данные
if data and 'response' in data:
    groups = data['response']['items']
    print(f"Найдено групп: {data['response']['count']}")
    
    # Cохраняем в DataFrame
    df = pd.DataFrame(groups)
    display(df)  # Отобразим таблицу
else:
    print("Ошибка в ответе API или нет данных.")
```

Задание по телеграмму (вывести чат)

```python
from telethon import TelegramClient
import asyncio  # Для async/await

# Данные для доступа к API telegamm
api_id = api_id2  # число
api_hash = api_hash2  # строка

async def main():
    async with TelegramClient('my_session', api_id, api_hash) as client:
        
        chat = '@gamediskont'  # Username чата
        
        print(f"Выводим последние 10 сообщений из чата {chat}:")
        
        async for message in client.iter_messages(chat, limit=10):
            sender = await message.get_sender()
            sender_name = sender.first_name if sender else 'Unknown'
            print(f"Отправитель: {sender_name}, Текст: {message.text or '[Медиа/Сервис]'}")
await main()
```

Задание по телеграмму (вывести каналы с количеством подпискивов более 200)

```python
async def main():
    async with TelegramClient('my_session', api_id, api_hash) as client:
        print("Доступные каналы с >200 подписчиками:")
        async for dialog in client.iter_dialogs():
            entity = dialog.entity
            # Фильтр: только каналы (broadcast=True), не супергруппы
            if hasattr(entity, 'broadcast') and entity.broadcast and \
               hasattr(entity, 'participants_count') and \
               entity.participants_count is not None and entity.participants_count > 200:
                print(f"Канал: {entity.title}, Подписчики: {entity.participants_count}")

await main()  
```
